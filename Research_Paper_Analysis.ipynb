#Generating the Embeddings for 15000

```python
#Embeddings-->Batch Size-->Ram Consumption Less
embedding=model.encode(df["paper_text"].to_list(),batch_size=32,show_progress_bar=True)
```

```python
print(embedding.shape)
print(type(embedding))
print(embedding.dtype)
```

FAISS --> It creates a special search structure

Cosine Similarity --> dot product hota hai

FAISS --> Inner product hota hai

L2 Normalization

```python
!pip install faiss-cpu
```

```python
import faiss
```

```python
# Normalise embedding
faiss.normalize_L2(embedding)
```

```python
# Create index
# Here IP means Inner Product it means it creates inner product for all embeddings
index=faiss.IndexFlatIP(384)
index
```

```python
# Add the L2 normalized embeddings to the FAISS index.
# This prepares the index for efficient similarity searches.
index.add(embedding)
```

```python
print(index.ntotal)
```

```python
# Create embedding for a query
query="deep learning for medical image analysis"
query_embedding=model.encode([query])
print(query_embedding.shape)
```

```python
# Normalize the query_embedding
faiss.normalize_L2(query_embedding)
```

```python
# To search how closely our query matches with the stored vectors of the research papers
D,I=index.search(query_embedding,5)
print(D) # Similarity Score --> tells us the similarity score between our query nd the reseach papers stored in the FAISS
print(I) # Indices --> It tells konse paper no seh it is similar
```
